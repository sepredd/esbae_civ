```python
import warnings
import numpy as np
from imblearn.ensemble import BalancedRandomForestClassifier
from imblearn import FunctionSampler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import IsolationForest
from skopt import BayesSearchCV
from sklearn.utils import class_weight, shuffle

def outlier_rejection(X, y):

    """This will be our function used to resample our dataset."""
    model = IsolationForest(max_samples=100, contamination=0.1)
    model.fit(X)
    y_pred = model.predict(X)
    return X[y_pred == 1], y[y_pred == 1]


def binary_probability_classification(
        df, y, predictors=None, class_prob=1, outlier=True, random=42, bayes=False
):
    
    randomness=7
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")

    warnings.filterwarnings(
        'ignore', 'This import path', FutureWarning
    )

    warnings.filterwarnings(
        'ignore', 'to make it possible to propagate', UserWarning
    )
    warnings.filterwarnings(action='once')
    # get target variable
    y_train = df[y][~df[y].isna()]
    #print(len(y_train[y_train == 1]))
    #print(len(y_train[y_train == 0]))
    if len(np.unique(y_train)) != 2:
        raise ValueError('Target variable y can only have 2 values')

    # prepare predictive variable
    x_df = df[predictors].copy() if predictors else df.copy()
    imp = SimpleImputer(strategy="mean")
    x_df = imp.fit_transform(x_df)

    x_train = x_df[~df[y].isna()]
    
    # compute class weights
    class_weights = class_weight.compute_class_weight(
        'balanced', 
        classes=np.unique(y_train), y=y_train
    )
    class_weight_dict = dict(enumerate(class_weights))
    
    if outlier:
        reject_sampler = FunctionSampler(func=outlier_rejection)
        x_train, y_train = reject_sampler.fit_resample(x_train, y_train)

    #print(len(y_train[y_train == 1]))
    #print(len(y_train[y_train == 0]))
    # model optimization
    parameters = {
        'n_estimators': [250, 500],
        'max_depth': [5, 10, 25, 50],
        'min_samples_split': [5, 10, 25],
        'min_samples_leaf': [10, 25]
    }

    if bayes:
        brf = BayesSearchCV(
            BalancedRandomForestClassifier(
                random_state=42, 
                oob_score=True,
                class_weight=class_weight_dict,
                n_jobs=-1
            ),
            search_spaces=parameters,
            n_iter=5,
            cv=5
        )
        # run classifier
        brf.fit(x_train, y_train)
        print(brf.best_estimator_)
        print(brf.best_score_)
        print(f'OOB Score is {brf.best_estimator_.oob_score_}')
        
    else:
        print('Generating the model by fitting predictors to y')
        brf = BalancedRandomForestClassifier(
            n_estimators=150, 
            random_state=randomness, 
            oob_score=True, 
            class_weight=class_weight_dict,
            n_jobs=-1
        )
        brf.fit(x_train, y_train)
        print(f'OOB Score is {brf.oob_score_}')


    # predict probability
    class_probabilities = brf.predict_proba(x_df)
    return class_probabilities.T[class_prob]    
```

```python
predictors= [
    'gfc_tc00', 'esa_lc21', 
    'lang_tree_height', 'potapov_tree_height', 
    'elevation', 'slope', 'aspect',
    'green_mean', 'green_sd',
    'red_mean', 'red_sd',
    'nir_mean', 'nir_sd',
    'swir1_mean', 'swir1_sd',
    'swir2_mean', 'swir2_sd',
    'ndfi_mean', 'ndfi_sd'
]


def get_stats(row, band, start, end):
    
    idx = [start*10000 < int(d) < end*10000 for d in row.dates]
    ts = np.array(row.ts[band])[idx]
    
    return np.nanmean(ts), np.nanstd(ts)

bands = joined_df.ts.head(1).values[0].keys()
start_year, end_year = 2015, 2022
for year in range(int(start_year), int(end_year)+1):
    if year == int(start_year) or year == int(end_year):
    #if year == int(end_year):
        
        # we create an "annual" dataframe for FNF classification of the current year
        to_class = joined_df[[
            'plot_id', 
            'gfc_tc00', 'esa_lc21', 'lang_tree_height', 'potapov_tree_height', 
            'elevation', 'slope', 'aspect', f'FNF_{year}'
        ]].copy()
        
        stats_list = []
        for band in bands:
            print(f'Calculating annual mean and sd for band {band} in year {year}')
            to_class[[f'{band}_mean', f'{band}_sd']] = joined_df.apply(lambda row: get_stats(row, band, int(year), int(year)+1), axis=1, result_type='expand')
            
        print(f'Calculating FNF classification probabilities for year {int(year)}')
        joined_df[f'fnf_prob_{year}'] = binary_probability_classification(
            to_class, 
            f'FNF_{year}', 
            predictors=predictors,
            outlier=True, 
            bayes=False
        )    
    #break
```

```python
from sampling_handler.sampling import sfc

def select_with_space_filling_curve(test_df, sample_df, strata_column='kmeans'):
    dfs = []
    for stratum in test_df[strata_column].unique():
        print(stratum)
        dfs.append(sfc.sfc_subsample(
            gdf = test_df[test_df[strata_column] == stratum],
            target_point_size = sample_df.neyman[sample_df.KMeans == stratum].values[0]
        ))
        
    return pd.concat(dfs)

samples = select_with_space_filling_curve(joined_df, sample_df) 
```
